Now we have a good estimate on how well each method performed. However, it is nice to have only one number representing the performance for each method and type of result (baseline and performance). We have the probability distribution of baseline and performance now, so it is sensible to compare them by their expected value. So, we have to compute the expected value for each baseline and performance distribution.

```python
import numpy as np

data = np.array([])
grouped = _5.groupby(level=['num_norm', 'num_oot', 'num_top', 'result'], axis=1)
for name, _ in grouped:
    pmf = _5[name].values
    supp = np.array(_5[name].columns)
    expected = np.sum(supp*pmf, axis=1)
    data = np.concatenate((data, expected))

index = _5.index.copy()
columns = pd.MultiIndex.from_tuples(columns, names=['num_norm', 'num_oot', 'num_top', 'result'])
m, n = len(index), len(columns)
data = data.reshape((n, m))
df2 = pd.DataFrame(data.T, index=index, columns=columns)
```
